Write a Python program that checks whether a given number is prime or not. 
A prime number is a natural number greater than 1 that has no positive divisors other than 1 and itself.

```python
num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(f"{num} is not a prime number")
            break
    else:
        print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message

```python
import random
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)
user_answer = int(input(f"What is the product of {num1} and {num2}? "))
correct_answer = num1 * num2
if user_answer == correct_answer:
    print("Correct! Well done.")
else:
    print(f"Incorrect. The correct answer is {correct_answer}.")
```

Exercise 4 Hospital Billing System

```python
room_charge_per_day = 2000
doctor_fee = 1500
lab_test_charge_per_test = 300
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

room_total = room_charge_per_day * days_stayed
lab_total = lab_test_charge_per_test * num_lab_tests
total_bill_before_discount = room_total + doctor_fee + lab_total + medicine_charges

if total_bill_before_discount > 10000:
    discount = 0.10 * total_bill_before_discount
else:
    discount = 0

final_bill = total_bill_before_discount - discount

print("===== Hospital Bill =====")
print(f"Room Charges ({days_stayed} days): ₹{room_total}")
print(f"Doctor Consultation Fee: ₹{doctor_fee}")
print(f"Lab Test Charges ({num_lab_tests} tests): ₹{lab_total}")
print(f"Medicine Charges: ₹{medicine_charges}")
print(f"Discount Applied: ₹{discount}")
print(f"Final Bill Amount: ₹{final_bill}")
```
