Handle missing values

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
```

```python
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
scaler_standard = StandardScaler()
scaler_minmax = MinMaxScaler()
```

```python
df_standard = df.copy()
df_minmax = df.copy()
df_standard[num_cols] = scaler_standard.fit_transform(df[num_cols])
df_minmax[num_cols] = scaler_minmax.fit_transform(df[num_cols])
```

Encoding Techniques

```python
small_cat_cols = [col for col in cat_cols if df[col].nunique() < 5]

df_encoded = pd.get_dummies(df, columns=small_cat_cols, drop_first=True)
df_encoded.head()
```

```python
large_cat_cols = [col for col in cat_cols if df[col].nunique() >= 5]

le = LabelEncoder()
for col in large_cat_cols:
    df_encoded[col] = le.fit_transform(df_encoded[col])

df_encoded.head()
```

FEATURE ENGINEERING

```python
df_encoded['age_group'] = pd.cut(
    df_encoded['age'],
    bins=[0, 25, 45, 65, 100],
    labels=['Young', 'Adult', 'Middle Aged', 'Senior']
)
```

```python
df_encoded['capital_ratio'] = df_encoded['capital_gain'] / (df_encoded['capital_loss'] + 1)

df_encoded[['age', 'age_group', 'capital_ratio']].head()
```

```python
df_encoded = pd.get_dummies(df_encoded, columns=['age_group'], drop_first=True)
df_encoded.head()
```

```python
df_encoded["capital_gain_log"] = np.log1p(df_encoded["capital_gain"])
df_encoded[["capital_gain", "capital_gain_log"]].head()
```
