```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('sales_data_with_discounts.csv.')
df.head()
```

```python
df.info()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_cols)
```

```python
df.describe()
```

```python
df.head()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col], color='lightblue')
    plt.title(f'Box Plot of {col}', fontsize=14)
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
date_columns = ['Date', 'City'] 
categorical_cols = [col for col in categorical_cols if col not in date_columns]

print("Categorical columns (excluding date):", categorical_cols)
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    order = df[col].value_counts().index  
    sns.countplot(data=df, y=col, order=order, palette="deep")  
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()
```
