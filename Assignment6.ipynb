Confidence Interval 

```python
import numpy as np
from scipy import stats
```

```python
data = np.array([1.13, 1.55, 1.43, 0.92, 1.25,
                 1.36, 1.32, 0.85, 1.07, 1.48,
                 1.20, 1.33, 1.18, 1.22, 1.29])
```

```python
n = len(data)  
n
```

```python
sample_mean = np.mean(data)  
sample_mean
```

```python
sample_sd = np.std(data, ddof=1)
sample_sd
```

1.Build 99% Confidence Interval Using Sample Standard Deviation

```python
confidence = 0.99
alpha = 1 - confidence
alpha
```

```python
df = n - 1 
df
```

```python
t_critical = stats.t.ppf(1 - alpha/2, df)
t_critical
```

```python
standard_error_t = sample_sd / np.sqrt(n)
standard_error_t
```

```python
margin_of_error_t = t_critical * standard_error_t
margin_of_error_t
```

```python
ci_lower_t = sample_mean - margin_of_error_t
ci_lower_t
```

```python
ci_upper_t = sample_mean + margin_of_error_t
ci_upper_t
```

2.Build 99% Confidence Interval Using Known Population Standard Deviation

```python
sigma = 0.2 
```

```python
z_critical = stats.norm.ppf(1 - alpha/2)
z_critical
```

```python
standard_error_z = sigma / np.sqrt(n)
standard_error_z
```

```python
margin_of_error_z = z_critical * standard_error_z
margin_of_error_z
```

```python
ci_lower_z = sample_mean - margin_of_error_z
ci_lower_z
```

```python
ci_upper_z = sample_mean + margin_of_error_z
ci_upper_z
```

Results

```python
print("============== FINAL RESULTS =================")
print(f"Sample Mean (x̄): {sample_mean:.4f}")
print(f"Sample Standard Deviation (s): {sample_sd:.4f}")
print()

print("A) 99% Confidence Interval using sample SD (t-distribution):")
print(f"   Lower Bound: {ci_lower_t:.4f}")
print(f"   Upper Bound: {ci_upper_t:.4f}")
print()

print("B) 99% Confidence Interval using known σ = 0.2 (z-distribution):")
print(f"   Lower Bound: {ci_lower_z:.4f}")
print(f"   Upper Bound: {ci_upper_z:.4f}")
print("=================================================")
```

HYPOTHESIS TESTING

```python
import math
```

```python
units_mean = 600                
units_sd = 25
```

```python
theoretical_mean_cost = 1000 + 5 * units_mean
theoretical_mean_cost
```

```python
sample_mean_cost = 3050          
sample_size = 25
```

```python
print("=== TASK 1: HYPOTHESIS STATEMENTS ===")
print(f"Theoretical Mean Cost (μ0) = {theoretical_mean_cost} Rs")
```

Calculate the Test Statistic

```python
population_sd_cost = 5 * units_sd
population_sd_cost
```

```python
standard_error_cost = population_sd_cost / math.sqrt(sample_size)
standard_error_cost
```

```python
z_statistic = (sample_mean_cost - theoretical_mean_cost) / standard_error_cost
z_statistic
```

```python
print("=== TASK 2: TEST STATISTIC ===")
print(f"Population SD of Cost (σ) = {population_sd_cost} Rs")
print(f"Standard Error (SE) = {standard_error_cost:.4f}")
print(f"Z-Test Statistic = {z_statistic:.4f}")
```

Determine the Probability and compare

```python
alpha = 0.05 
```

```python
critical_z_value = stats.norm.ppf(1 - alpha)
critical_z_value
```

```python
p_value = 1 - stats.norm.cdf(z_statistic)
p_value
```

```python
print("=== TASK 3: P-VALUE & CRITICAL VALUE ===")
print(f"Significance Level (α) = {alpha}")
print(f"Critical Z-Value = {critical_z_value:.4f}")
print(f"P-Value = {p_value:.6f}")
```

Make the Decision

```python
if z_statistic > critical_z_value:
    print("Decision: Reject H₀")
else:
    print("Decision: Fail to Reject H₀")
```
