Question1

```python
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

numbers.append(60)
print("After append(60):", numbers)

numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

numbers.insert(2, 25)  
print("After insert(2, 25):", numbers)

numbers.remove(40)
print("After remove(40):", numbers)

popped_value = numbers.pop()  
print("After pop():", numbers, "| Popped value:", popped_value)

temp_list = numbers.copy()  
temp_list.clear()
print("After clear():", temp_list)

index_30 = numbers.index(30)
print("Index of 30:", index_30)

count_20 = numbers.count(20)
print("Count of 20:", count_20)

numbers.sort()
print("After sort():", numbers)

numbers.reverse()
print("After reverse():", numbers)
```

```python
marks = 82
if marks >= 90:
    grade = "A"
elif 75 <= marks <= 89:
    grade = "B"
elif 50 <= marks <= 74:
    grade = "C"
else:
    grade = "F"  
print(f"Marks: {marks} | Grade: {grade}")
```

```python
num = int(input("Enter a number: "))
if num > 0:
    print("The number is positive.")
    if num % 2 == 0:
        print("It is also even.")
    else:
        print("It is odd.")
elif num == 0:
    print("The number is zero.")
else:
    print("The number is negative.")
```

```python
import pandas as pd
import numpy as np
data = {
    'Name': ['Pratik', 'Raj', 'Shyam', 'Ram', 'Sarthak', 'Suhel', 'Darshan', 'Prashu', 'Dhairya', 'Aditya'],
    'Age': [25, 30, 35, 40, 28, 33, 29, 31, 27, 26],
    'City': ['Mumbai', 'Delhi', 'Bangalore', 'Chennai', 'Mumbai', 'Delhi', 'Kolkata', 'Chennai', 'Bangalore', 'Delhi'],
    'Salary': [50000, 60000, 70000, 80000, np.nan, 65000, 72000, 75000, 68000, 60000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```
